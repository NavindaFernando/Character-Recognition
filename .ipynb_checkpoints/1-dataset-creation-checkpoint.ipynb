```python
import os
import numpy as np

# import dataset 
data_path='dataset'

# The file names in the dataset are added to an array list
categories = os.listdir(data_path)
print(categories)

# Numbers are generated according to the number of files.
labels = [i for i in range(len(categories))]
print(labels)

# An array dictionary is created using file names (categories) and numbers (labels).
category_dict = {'a':0,'ae':1,'e':2,'u':3}

print(category_dict)
```

```python
import cv2

data = []
target = []

# go to all files for one by one
for category in categories:
    imgs_path=os.path.join(data_path,category) # Create a new path by using data_path (dataset) and filenames (categories)
    print(imgs_path)
    img_names=os.listdir(imgs_path) # The names of the images inside the created path are added to an array list
    print(img_names)
    
    # go to all images for one by one
    for img_name in img_names:
        img_path = os.path.join(imgs_path,img_name) # Create a full path by using data_path (dataset) and categories (filenames) and  img_path (imagenames)
        print(img_path)
        img = cv2.imread(img_path,0) # load to gray image
        img = cv2.resize(img,(8,8)) # convert to 8,8 image
        data.append(img) # append resized images to data=[]
        target.append(category_dict[category]) # append target to target=[]
        
    print('---------------------------------------------')
```

```python
print(len(data),len(target)) # The total number of images in the dataset
```

```python
from matplotlib import pyplot as plt

# show 8,8 image (10)
plt.imshow(data[10],cmap='gray')
```

```python
# convert data to numpy array
data = np.array(data)
print('before resize:',data.shape)

# convert target to numpy array
target = np.array(target)
print('before resize:',target.shape)
```

```python
# conver to 1D array (flattening)
data = data.reshape(data.shape[0],data.shape[1]*data.shape[2])

print('after resize:',data.shape)
print('after resize:',target.shape)

# This is saved as a numpy array to a physical file
np.save('data',data) 
np.save('target',target)
```
